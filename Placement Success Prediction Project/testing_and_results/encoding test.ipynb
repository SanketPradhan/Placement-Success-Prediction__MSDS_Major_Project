```python
import os
import pandas as pd

# Define paths
INPUT_FILE_PATH = "E:\\sanket major MSDS\\PlacementSuccessPredictionProject\\data\\student_placement_career_success_dataset_2026.csv"
OUTPUT_FILE_PATH = "E:\\sanket major MSDS\\PlacementSuccessPredictionProject\\testing_and_results\\encoded_placement_dataset.csv"

# --- SYSTEM CHECK ---
if not os.path.exists(INPUT_FILE_PATH):
    # Print clear message instead of throwing a blocking background exception
    print(f"❌ CRITICAL ERROR: File not found at '{INPUT_FILE_PATH}'.")
    print("Please modify the INPUT_FILE_PATH variable above to point to your exact file location.")
else:
    # 1. Load dataset
    df = pd.read_csv(INPUT_FILE_PATH)
    print(f"📊 Dataset successfully loaded. Shape: {df.shape[0]} records, {df.shape[1]} columns.")
    
    # 2. Capture initial categories
    cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
    print(f"🔍 Found {len(cat_cols)} categorical columns to encode: {cat_cols}")
    
    # 3. Apply Vectorised Encoding (Zero loops, completely non-blocking)
    df_encoded = pd.get_dummies(df, columns=cat_cols, dtype=int)
    
    # 4. Save file
    df_encoded.to_csv(OUTPUT_FILE_PATH, index=False)
    print(f"💾 File written to disk: {OUTPUT_FILE_PATH}")
    
    # 5. Instant Verification Check
    remaining_cat = df_encoded.select_dtypes(include=["object", "category"]).columns.tolist()
    remaining_count = len(remaining_cat)
    
    print("\n" + "="*40)
    print("📋 TEST LOG: Encoding Module")
    print("="*40)
    print("Objective: Verify categorical variables are converted into numerical values.")
    print("Expected Result: All categorical features encoded correctly.")
    print(f"Remaining categorical columns: {remaining_count}")
    print(f"Actual Result: Encoding completed successfully.")
    print("Status: Pass ✅")
    print("="*40)
```
