```python
# Step 1: Import required libraries
import os
import pandas as pd

# Step 2: Define your dataset paths
INPUT_FILE_PATH = "E:\\sanket major MSDS\\PlacementSuccessPredictionProject\\data\\student_placement_career_success_dataset_2026.csv"
OUTPUT_FILE_PATH = "E:\\sanket major MSDS\\PlacementSuccessPredictionProject\\testing_and_results\\duplicated_placement_dataset.csv"

# Step 3: Load the dataset
if not os.path.exists(INPUT_FILE_PATH):
    raise FileNotFoundError(
        f"Could not find the dataset at: {INPUT_FILE_PATH}. Please check your path string."
    )

df = pd.read_csv(INPUT_FILE_PATH)
print(f"Dataset loaded. Initial shape: {df.shape[0]} records, {df.shape[1]} columns.")

# Calculate initial duplicates
initial_duplicate_count = df.duplicated().sum()
print(f"Total duplicate records found: {initial_duplicate_count}\n")


# Step 4: Duplicate Removal Module
def remove_duplicate_records(input_df):
    """Identifies and eliminates completely duplicated rows, keeping the first occurrence."""
    processed_df = input_df.copy()

    # Drop duplicate rows across all columns
    # keep='first' ensures the original record is preserved
    processed_df = processed_df.drop_duplicates(keep="first")

    return processed_df


# Step 5: Execute the deduplication module
print("--- Starting Duplicate Removal Module ---")
df_cleaned = remove_duplicate_records(df)
print("------------------------------------------\n")


# Step 6: Export Deduplicated Data
df_cleaned.to_csv(OUTPUT_FILE_PATH, index=False)
print(f"Deduplicated dataset saved to: {OUTPUT_FILE_PATH}")
print(f"Final shape: {df_cleaned.shape[0]} records.")


# Step 7: Automated Test Verification Block
remaining_duplicates = df_cleaned.duplicated().sum()

print("\n--- Test Verification Result ---")
print(f"Module Name: Duplicate Removal Module")
print(f"Objective: Ensure duplicate student records are removed.")
print(f"Expected Result: Duplicate entries eliminated.")
print(f"Remaining duplicate entries in dataset: {remaining_duplicates}")

if remaining_duplicates == 0:
    print("Actual Result: Duplicates removed successfully.")
    print("Status: Pass ✅")
else:
    print("Actual Result: Duplicate records still exist.")
    print("Status: Fail ❌")
```
