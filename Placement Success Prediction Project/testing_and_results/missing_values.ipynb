```python
# Step 1: Import required libraries
import os
import pandas as pd

# Step 2: Define your dataset paths
# REPLACE the string below with your actual file path (e.g., "C:/Users/Name/Downloads/dataset.csv")
INPUT_FILE_PATH = "E:\\sanket major MSDS\\PlacementSuccessPredictionProject\\data\\student_placement_career_success_dataset_2026.csv"
OUTPUT_FILE_PATH = "E:\\sanket major MSDS\\PlacementSuccessPredictionProject\\testing_and_results\\tested_missing_value_data.csv"

# Step 3: Load the dataset
if not os.path.exists(INPUT_FILE_PATH):
    raise FileNotFoundError(
        f"Could not find the dataset at: {INPUT_FILE_PATH}. Please check your path string."
    )

df = pd.read_csv(INPUT_FILE_PATH)
print(f"Dataset successfully loaded. Shape: {df.shape[0]} rows, {df.shape[1]} columns.")

initial_missing_count = df.isnull().sum().sum()
print(f"Total missing values found across all columns: {initial_missing_count}\n")


# Step 4: Missing Value Handling Module
def handle_missing_values(input_df):
    """Processes numerical missing values using mean and categorical values using mode."""
    processed_df = input_df.copy()

    for column in processed_df.columns:
        if processed_df[column].isnull().any():
            # Check if column data type is numerical
            if pd.api.types.is_numeric_dtype(processed_df[column]):
                mean_value = processed_df[column].mean()
                processed_df[column] = processed_df[column].fillna(mean_value)
                print(
                    f"Filled missing values in numerical '{column}' with mean: {mean_value:.2f}"
                )

            # Column is categorical (Object/String types)
            else:
                # Extracts the primary mode value if multiple modes exist
                mode_value = processed_df[column].mode()[0]
                processed_df[column] = processed_df[column].fillna(mode_value)
                print(
                    f"Filled missing values in categorical '{column}' with mode: '{mode_value}'"
                )

    return processed_df


# Step 5: Execute the processing module
print("--- Starting Missing Value Processing Module ---")
df_cleaned = handle_missing_values(df)
print("------------------------------------------------\n")


# Step 6: Export Cleaned Data
df_cleaned.to_csv(OUTPUT_FILE_PATH, index=False)
print(f"Cleaned dataset successfully saved to: {OUTPUT_FILE_PATH}")


# Step 7: Automated Test Verification Block
remaining_missing = df_cleaned.isnull().sum().sum()

print("\n--- Test Verification Result ---")
print(f"Objective: Verify correct handling of missing values.")
print(
    f"Expected Result: Missing numerical values replaced using mean/median and categorical values using mode."
)
print(f"Remaining missing values in dataset: {remaining_missing}")

if remaining_missing == 0:
    print("Actual Result: Missing values processed successfully. PASSED ✅")
else:
    print("Actual Result: Missing values still exist. FAILED ❌")
```
